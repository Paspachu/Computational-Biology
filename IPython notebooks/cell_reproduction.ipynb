First, import packages, and give their contents a prefix name (that's the "as").  We'll always do this. 

```python
import matplotlib.pylab as plt   # That gives plotting, and the next line makes plots appear inline in notebook
%matplotlib inline  
import numpy as np  # That gives numerical arrays 
```

To start, we define an numpy array corresponding to our variable $n(t)$; this is what we operate on when numerically simulating this model. 

In particular, we define an array `nArray` with `numGen` entries (all zeros, for now), where `numGen` is the number of generations over which we simulate this model, including generation 0. 

Then the $i$th entry of `nArray` stores the number of cells at generation $i$.

```python
numGen=20;  # total number of generations over which to simulate dynamical model
nArray=np.zeros(numGen) # initialize nArray with all zeros
print(nArray) # use print to display a "variable"
```

Initialize the first entry in `nArray` to the number of individuals at generation 0.  

Note:  Python uses `[]` to index into arrays, and INDEXING STARTS AT 0 (not 1).

```python
nArray[0]=1000
```

We'll make a `for` loop, to run our dynamical model.  

First, we need to define what we'll be looping over: the iterator tArray, which counts which generation we are on.  

We construct it using `np.arange(start,stop)`, where start and stop are two integers, makes list from `start` to `stop-1`

```python
tArray=np.arange(0,numGen-1)
print(tArray)
```

Finally, let us define the model parameter $p$ (fraction of cells that survive each division).

```python
p=0.95; #fraction of cells that survive each reproduction
```

## Run the dynamical model!

```python
for t in tArray:
    nArray[t+1]=4*nArray[t]*p
```

Plot!

```python
plt.plot(nArray,'ko',markersize=5)
plt.xlabel('generation')
plt.ylabel('number of cells')
plt.title('cell reproduction (linear model)')
```

### Now move to the nonlinear cell reproduction model

```python
numGen=200  # total number of generations for simulation

nArray=np.zeros(numGen) # array of number of cells at each generation
nArray[0]=1000

tArray=np.arange(0,numGen-1)

# parameters: p and K
p=0.95 #fraction of cells that survive each reproduction
K=2000  #reproduction capacity
```

## Run the dynamical model!

```python
for t in tArray:
    nArray[t+1]=4*nArray[t]*p*(1-nArray[t]/K) 
```

```python
plt.plot(nArray,'ko',markersize=5)
plt.xlabel('generation')
plt.ylabel('number of cells')
plt.title('cell reproduction (nonlinear model)')
```
