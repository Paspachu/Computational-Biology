## Analysis of current data

```python
import matplotlib.pylab as plt
%matplotlib inline  
import numpy as np
```

#### Load current data from a text file

Data are succesive measurements of current in a population of ion channels (time series - current at different time points)

```python
data = np.loadtxt('CurrentsData.dat')
```

```python
print(data.shape)
data[0:10,0:5]
```

#### Analyze statistics:

Plot mean versus variance, get estimates of $N$ (the number of channels) and $i$ (the current through a single open channel).

```python
E = np.mean(data, axis=1)
Var = np.var(data, axis=1)

p = np.polyfit(E, Var, 2)
print(p)
plt.scatter(E, Var)
plt.plot(E, p[0]*E**2 + p[1]*E + p[2], 'red')
plt.show()
```

```python
N = -1 / p[0]
i = p[1]
print(N, i)
```
