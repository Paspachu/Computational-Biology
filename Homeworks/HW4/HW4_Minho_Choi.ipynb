### Problem 1

We claim that the dwell time distribution is sum of 4 exponenetial random variables:
$$\text{P}(\text{dwell time in } \{O_1, O_2, O_3, O_4\} = k \text{ steps}) = c_1 \lambda^k_1 + c_2 \lambda^k_2 + c_3 \lambda^k_3 + c_4 \lambda^k_4$$
To prove the claim, we first identify the state space we are working on, which is:
$$\Omega = \{O_1, O_2, O_3, O_4, C_1, C_2\} = \{S_1, S_2, S_3, S_4, S_5, S_6\}$$
Let the transition matrix $A$ be:
$$A = \begin{pmatrix}
a_{11} & a_{12} & a_{13} & a_{14} & a_{15} & a_{16} \\
a_{21} & a_{22} & a_{23} & a_{24} & a_{25} & a_{26} \\
a_{31} & a_{32} & a_{33} & a_{34} & a_{35} & a_{36} \\
a_{41} & a_{42} & a_{43} & a_{44} & a_{45} & a_{46} \\
a_{51} & a_{52} & a_{53} & a_{54} & a_{55} & a_{56} \\
a_{61} & a_{62} & a_{63} & a_{64} & a_{65} & a_{66}
\end{pmatrix}$$
Since Markov Chain is time-homogeneous, we have to show that our claim is true for time step $k = 0$. Then, the claim is true for any time step $k$. Hence, we assume that our state at time step $k=0$ is open:
$$X_0 \in \{O_1, O_2, O_3, O_4\}$$
Then, we know that:
$$\vec{p}(0) = \begin{pmatrix}
p_1(0) \\
p_2(0) \\
p_3(0) \\
p_4(0) \\
0 \\
0
\end{pmatrix}$$
By the definition of Markov Chain, we get:
$$\begin{pmatrix}
p_1(1) \\
p_2(1) \\
p_3(1) \\
p_4(1) \\
p_5(1) \\
p_6(1)
\end{pmatrix} = \begin{pmatrix}
a_{11} & a_{12} & a_{13} & a_{14} & a_{15} & a_{16} \\
a_{21} & a_{22} & a_{23} & a_{24} & a_{25} & a_{26} \\
a_{31} & a_{32} & a_{33} & a_{34} & a_{35} & a_{36} \\
a_{41} & a_{42} & a_{43} & a_{44} & a_{45} & a_{46} \\
a_{51} & a_{52} & a_{53} & a_{54} & a_{55} & a_{56} \\
a_{61} & a_{62} & a_{63} & a_{64} & a_{65} & a_{66}
\end{pmatrix} \begin{pmatrix}
p_1(0) \\
p_2(0) \\
p_3(0) \\
p_4(0) \\
0 \\
0
\end{pmatrix}$$
Since we are only interested in transition between open states, we only consider the first 4 rows of $A$. Also, last two entries of $\vec{p}(0)$ are zeros, so the last two columns of $A$ are eliminated. Hence, we only consider the uppper 4 by 4 corner part of $A$:
$$\begin{pmatrix}
p_1(1) \\
p_2(1) \\
p_3(1) \\
p_4(1) \\
\end{pmatrix} = \begin{pmatrix}
a_{11} & a_{12} & a_{13} & a_{14} \\
a_{21} & a_{22} & a_{23} & a_{24} \\
a_{31} & a_{32} & a_{33} & a_{34} \\
a_{41} & a_{42} & a_{43} & a_{44} 
\end{pmatrix} \begin{pmatrix}
p_1(0) \\
p_2(0) \\
p_3(0) \\
p_4(0) \\
\end{pmatrix}$$
Here, we define:
$$p^{(c)}_i(k) = \text{P}(X_k = S_i | X_l = \{S_1, S_2, S_3, S_4\} \text{ for all } 0 \leq l \leq k-1)$$
where $i = 1, 2, 3, 4$. Then, for $k \geq 1$, we get:
$$\begin{pmatrix}
p^{(c)}_1(k+1) \\
p^{(c)}_2(k+1) \\
p^{(c)}_3(k+1) \\
p^{(c)}_4(k+1) \\
\end{pmatrix} = \begin{pmatrix}
a_{11} & a_{12} & a_{13} & a_{14} \\
a_{21} & a_{22} & a_{23} & a_{24} \\
a_{31} & a_{32} & a_{33} & a_{34} \\
a_{41} & a_{42} & a_{43} & a_{44} 
\end{pmatrix} \begin{pmatrix}
p^{(c)}_1(k) \\
p^{(c)}_2(k) \\
p^{(c)}_3(k) \\
p^{(c)}_4(k) \\
\end{pmatrix}$$
$$\Rightarrow \vec{p}^{(c)}(k+1) = \tilde{A} \; \vec{p}^{(c)}(k)$$
We know that the solution to the above linear map is:
$$\vec{p}^{(c)}(k) = k_1\vec{v}^{(1)}\lambda^k_1 + k_2\vec{v}^{(2)}\lambda^k_2 + k_3\vec{v}^{(3)}\lambda^k_3 + k_4\vec{v}^{(4)}\lambda^k_4$$
where $\lambda_i, \vec{v}^{(i)} \text{ for } i = 1, 2, 3, 4$ are the eigenvalue-eigenvector pairs of $\tilde{A}$, and $k_1, k_2, k_3, k_4$ are arbitrary constants. Then, we get:
$$\begin{equation*}
\begin{split}
    \text{P}(&\text{channel remains open for at least } k \text{ steps}) = p^{(c)}_1(k) + p^{(c)}_2(k) + p^{(c)}_3(k) + p^{(c)}_4(k) \\
    &= k_1(v^{(1)}_1 + v^{(1)}_2 + v^{(1)}_3 + v^{(1)}_4)\lambda^k_1 + k_2(v^{(2)}_1 + v^{(2)}_2 + v^{(2)}_3 + v^{(2)}_4)\lambda^k_2 \\
    &\;\;\;+ k_3(v^{(3)}_1 + v^{(3)}_2 + v^{(3)}_3 + v^{(3)}_4)\lambda^k_3 + k_4(v^{(4)}_1 + v^{(4)}_2 + v^{(4)}_3 + v^{(4)}_4)\lambda^k_4 \\
    &:= \tilde{c}_1\lambda^k_1 + \tilde{c}_2\lambda^k_2 + \tilde{c}_3\lambda^k_3 + \tilde{c}_4\lambda^k_4
\end{split}
\end{equation*}$$
Hence, the dwell time in open states $T_o$ is:
$$\begin{equation*}
\begin{split}
    \text{P}(T_o \geq k) &= \text{P}(\text{channel remains opened for at least } k-1 \text{ steps}) \\
    &= \tilde{c}_1\lambda^{k-1}_1 + \tilde{c}_2\lambda^{k-1}_2 + \tilde{c}_3\lambda^{k-1}_3 + \tilde{c}_4\lambda^{k-1}_4
\end{split}
\end{equation*}$$
Note that it is $k-1$ steps, since we are in open state at time step $k=0$. \
Thus, we get:
$$\begin{equation*}
\begin{split}
    \text{P}(T_o = k) &= \text{P}(T_o \geq k) - \text{P}(T_o \geq k+1) \\ 
    &= \tilde{c}_1\lambda^{k-1}_1 + \tilde{c}_2\lambda^{k-1}_2 + \tilde{c}_3\lambda^{k-1}_3 + \tilde{c}_4\lambda^{k-1}_4 - \tilde{c}_1\lambda^k_1 - \tilde{c}_2\lambda^k_2 - \tilde{c}_3\lambda^k_3 - \tilde{c}_4\lambda^k_4 \\
    &:= c_1\lambda^k_1 + c_2\lambda^k_2 + c_3\lambda^k_3 + c_4\lambda^k_4
\end{split}
\end{equation*}$$

### Problem 2

Here we import all the packages we needed:

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline 
```

First, we initialize random number generator:

```python
rng = np.random.default_rng()
```

Also, we define the transition matrix:

```python
A = np.array([[0.98, 0.1, 0],
             [0.02, 0.7, 0.05],
             [0, 0.2, 0.95]])
A
```

We initialize maximum time step and states at each time step. We assume that we start from $S_1$ which is closed: 

```python
Tmax = 1000000
states = np.zeros(Tmax+1, dtype=int)
states[0] = 1
```

We run the simulation:

```python
for t in range(Tmax):
    r = rng.uniform(0,1,1)

    if r < A[0, states[t]-1]:
        states[t+1] = 1
    elif r >= A[0, states[t]-1] and r < A[0, states[t]-1] + A[1, states[t]-1]:
        states[t+1] = 2
    else:
        states[t+1] = 3
```

We create the reduced list that combine the close states together (closed state = 0 and open state = 1):

```python
rstates = np.zeros(len(states), dtype=int)
for (i, s) in enumerate(states):
    if s < 3:
        rstates[i] = 0
    else:
        rstates[i] = 1
```

Now, we calculate the dwell times in the closed state:

```python
state_changes = np.where(rstates[:-1] != rstates[1:])[0]
dwell_times = np.array([state_changes[0] + 1])

for i in range(2, len(state_changes), 2):
    dwell_times = np.append(dwell_times, state_changes[i] - state_changes[i-1] + 1)
```

The histogram of the dwell times looks like the following:

```python
plt.hist(dwell_times, 31)
plt.xlabel('Dwell times')
plt.ylabel('Frequency')
plt.title('Histogram of Dwell Times in Closed State')
plt.show()
```

To see if the histogram is fit by a single exponential, we find a best fit line to the log scale of frequency:

```python
freq, bins = np.histogram(dwell_times, bins=np.linspace(0, np.max(dwell_times), 31))
freq_modified = freq.copy()
freq_modified[freq_modified == 0] = 1
p = np.polyfit(bins[:-1], np.log(freq_modified), 1)
```

Then, the histogram of log scale of frequency with the best fit line is the following:

```python
x = np.linspace(0, bins[-1], 1000)
y = p[0] * x + p[1]
y[y < 0] = 0

plt.stairs(np.log(freq_modified), bins, fill=True)
plt.plot(x, y)
plt.xlabel('Dwell times')
plt.ylabel('Log of Frequency')
plt.title('Histogram of Log of Dwell Times in Closed State')
plt.show()
```

Without the log, the histogram with the best fit exponential line is the following:

```python
plt.hist(dwell_times, 31)
plt.plot(x, np.exp(p[1]) * np.exp(-np.exp(p[0]) * x / 100), label='Fit Line')
plt.plot(x, np.max(freq) * np.exp(-np.exp(p[0]) * x / 100), label='Fit Line with Scaling')
plt.xlabel('Dwell times')
plt.ylabel('Frequency')
plt.title('Histogram of Dwell Times in Closed State')
plt.legend()
plt.show()
```

As we can see from the above plot, the single exponential does not fit the histogram well. The spike in the first bin is too sharp for the single exponential to properly fit the histogram. The first bin has extremely high frequency compare to other bins because the probability of staying at state 2 which is $C_2$ is low relative to the probability of staying at states 1 and 3, $C_1$ and $C_3$ respectively. Moreover, the probability of moving from state 2 to state 3 is greater than the probability of moving from state 2 to state 1, so the dwell times in closed states are less than 30 most cases. Thus, from the plot, we can conclude there are more than one closed states in our membrane current.
