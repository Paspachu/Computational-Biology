Here we import all the packages we needed:

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline 

import scipy.linalg as la
```

From HW2, we get the projection matrix:

```python
A = np.zeros((51, 51))

for i in range(3, 50):
    A[0, i] = 0.24
    A[i+1, i] = 0.942

A[1, 0] = 1
A[2, 1] = 1
A[3, 2] = 0.0722
A[0, -1] = 0.24

np.set_printoptions(precision=3,suppress=True)
A
```

To compute the matrix of elasicities, we first compute the right and left eigenvectors:

```python
# Right eigenvectors
l,v = la.eig(A)
idx=np.argsort(np.abs(l))
l_sorted=l[idx]  
v_sorted=v[:,idx]
lambda_max= l_sorted[-1]
v_max = v_sorted[:,-1]

# Left eigenvectors
l,w = la.eig(A.T)
idx=np.argsort(np.abs(l))
l_sorted=l[idx] 
w_sorted= w[:,idx] 
w_max = w_sorted[:,-1]
```

Then, we compute the matrix of sensitivities:

```python
s = np.outer(w_max,v_max)/np.dot(w_max,v_max)

np.set_printoptions(precision=3,suppress=True)
print(np.real(s))
```

Now, we compute the matrix of elasticities:

```python
e= (s * A) / lambda_max
np.set_printoptions(threshold=np.inf)
print(np.real(e))
```

The first row of the above matrix shows the elasticities of fecundity values. The elasticities for fecundity values $f_a$ are the same for all ages $a \geq 3$. This makes sense as the fecundity values are same for all ages $a \geq 3$. The elasticities for $f_a$ where $a < 3$ are all zeros as the fecundity values are zeros for $a < 3$. In contrast, the elasticities for annual survival probabilities values $p_a$ decrease as the age value $a$ increases. Hence, the survival probabilities from birth to age 3 has the highest elasticities of 0.037. Since the survival probabilities from birth to age 3 are small relative to the survival probabilities of adults ($a \geq 3$), managing the survival probabilities from birth to age 3 contributes the most to the long-term population growth and population survival. Thus, for management plans, there should be some restrictions on the logging companies in harvesting the forests to save the owls' inhabitants. Moreover, some organizations should assist the infant owls under age 3 to increase their survival probabilities.

### Problem 2

First, we initialize the projection matrix from Exercise 2.13:

```python
A = np.array([[0, .0043, .1132, 0], [.9775, .9111, 0, 0], [0, .0736, .9534, 0], [0, 0, 0.0452, 0.9804]])
np.set_printoptions(precision=4,suppress=True)
A
```

Then, we compute the growth rate $\lambda$:

```python
l,v = la.eig(A)
idx=np.argsort(np.abs(l))
l_sorted=l[idx]
v_sorted=v[:,idx] 
lambda_max= l_sorted[-1]
print("Growth rate is: {}".format(np.real(lambda_max)))
```

Here, we can see that the growth rate is greater than 1, so the population of the killer whales will increase exponentially in the long-term. Next, we obtain the stable stage distribution with a total of 250 individuals:

```python
stableStage = v_sorted[:,-1]
stable_dist = np.array([stage / sum(stableStage) * 250 for stage in stableStage])
print("Stable stage distribution is: {}".format(np.round(stable_dist)))
```

Note that all elements in the stable stage distribution are rounded as the number of individuals must be integer (we still get the same answer if we do not round). Now, using the stable stage distrbution as the initial population structure, we find the maximum number of juveniles (or reproductive adults) that can be harvested without resulting extinction of the killer whales. We begin with creating a function that iterates the dynamical model and returns the growth rate $\lambda$:

```python
nZero = np.round(stable_dist)

def iterate_dynamical_model(nZero, H):
    Tmax = 200
    nT = np.zeros([4,Tmax])
    nT[:,0] = nZero

    for t in range(Tmax-1):
        nT[:,t+1] = np.dot(A,nT[:,t]) - H
        nT[:,t+1] = np.clip(nT[:,t+1], 0, None) # Change any negative value to 0    

    generations = np.arange(Tmax)
    pf = np.polyfit(generations, np.log(nT.sum(axis = 0)), 1)
    lambdaEst = np.exp(pf[0])

    return lambdaEst
```

To find the maximum number of juveniles that can be harvested, we compute:

```python
growth_rate = iterate_dynamical_model(nZero, np.array([0, 0, 0, 0]))
i = 0

while growth_rate > 1:
    i = i + 1
    growth_rate = iterate_dynamical_model(nZero, np.array([0, i, 0, 0]))

print("Maximum number of juveniles that can be harvested is: {}".format(i - 1))
```

Hence, the maximum number of juveniles that can be harvested is 5 without without resulting extinction of the killer whales. Next, we find the maximum number of reproductive adults that can be harvested:

```python
growth_rate = iterate_dynamical_model(nZero, np.array([0, 0, 0, 0]))
j = 0

while growth_rate > 1:
    j = j + 1
    growth_rate = iterate_dynamical_model(nZero, np.array([0, 0, j, 0]))

print("Maximum number of reproductive adults that can be harvested is: {}".format(j - 1))
```

Thus, the maximum number of reproductive adults that can be harvested is 3 without without resulting extinction of the killer whales.
